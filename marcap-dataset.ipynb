```python
import pandas as pd
import numpy as np
from marcap import marcap_data
```

```python
df = marcap_data('1996-01-01', '2020-10-15')
```

```python
df.head()
```

```python
df.tail()
```

```python
len(df)
```

```python
# 필요한 정보는 날짜, 종명이름, 시가총액, 시가총액 순위(당일) - 15위 안 데이터만 추출
# rank 15위 안에 데이터만 뽑기

df[df['Rank'] <= 15.0].head(20)
```

```python
df_15 = df[df['Rank'] <= 15.0]
```

```python
df_15 = df_15.reset_index(inplace=False)
df_15.head()
```

```python
df_15 = df_15[['Date', 'Name', 'Marcap']]
df_15.head()
```

```python
df_15.tail()
```

```python
df_15.to_csv('시가총액(1996-01-01 ~ 2020-10-15).csv', index=False)
```
